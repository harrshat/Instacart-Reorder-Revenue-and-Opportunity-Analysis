# Phase 2 — Segment Customers

```python
orders.sort_values(['user_id','order_number']).head(10)
```

```python
orders['avg_days_bw_orders'] = round(orders.groupby('user_id')['days_since_prior_order'].transform('median'),0)
orders
```

```python
orders.head(15)
```

```python
# Average days between orders for each user
order_cadence = orders[['user_id', 'avg_days_bw_orders']].drop_duplicates()
```

```python
order_cadence.sort_values('user_id')
```

```python
basket_size_perorder = orders.merge(orders_prior, how='inner', on='order_id').groupby(['user_id', 'order_id'])['product_id'].count().reset_index()
```

```python
basket_size_perorder.rename(columns = {"product_id":"basket_size"}).sort_values('user_id').head(15)
```

```python
avg_basket_size = round(basket_size_perorder.groupby('user_id')['product_id'].median().reset_index().rename(columns={'product_id': 'avg_basket_size'}),0)
avg_basket_size.head(5)
```

Basket size trend:

```python
orders.sort_values(['user_id','order_number'])[['order_id','user_id','order_number']]
```

```python
basket_size_perorder = orders_prior.groupby('order_id')['product_id'].nunique().reset_index().rename(columns = {"product_id":"basket_size"})
```

```python
basket_size_perorder.head()
```

```python
first2_orders = round(
    orders[orders['order_number'] <= 2]
    .merge(basket_size_perorder, how="left", on="order_id")
    .groupby('user_id')['basket_size'].median()
    .reset_index(), 0
)
```

```python
first2_orders
```

```python
first2_orders.rename(columns={"basket_size":"avg_basketsize_first2orders"})
```

```python
last2_orders = round((basket_size_perorder
    .merge(orders, how='right', on='order_id')
    .groupby('user_id')
    .apply(lambda x: x.nlargest(2, 'order_number')['basket_size'].median())
    .reset_index()
    .rename(columns={0: 'avg_basket_size_last2_orders'})
),0)
```

```python
last2_orders
```

```python
final = (order_cadence
    .merge(avg_basket_size, how='left', on='user_id')
    .merge(first2_orders, how='left', on='user_id')
    .merge(last2_orders, how='left', on='user_id')
)
```

```python
final.shape
```

```python
final.head()
```

```python
final.rename(columns={"basket_size":"avg_basket_size_first2_orders"},inplace=True)
```

```python
final
```

```python
final[final['avg_basket_size_first2_orders']>final['avg_basket_size_last2_orders']]
```

```python
final[final['avg_basket_size_first2_orders']<final['avg_basket_size_last2_orders']]
```

```python
final[final['avg_basket_size_first2_orders']==final['avg_basket_size_last2_orders']]
```

```python
conditions = [
    final['avg_basket_size_first2_orders'] > final['avg_basket_size_last2_orders'],
    final['avg_basket_size_first2_orders'] == final['avg_basket_size_last2_orders'],
    final['avg_basket_size_first2_orders'] < final['avg_basket_size_last2_orders']
]

choices = ['Declining basket size', 'Same Basket size', 'Increasing basket size']

final['basket_size_trend'] = np.select(conditions, choices, default='Incomplete data')
```

```python
final['basket_size_trend'].value_counts()
```

```python
final['basket_size_trend'].value_counts(normalize=True) * 100
```

```python
conditions = [
    (final['avg_days_bw_orders'] < 10) & (final['basket_size_trend'].isin(['Increasing basket size', 'Same Basket size'])),
    (final['avg_days_bw_orders'] > 20) | (final['basket_size_trend'] == 'Declining basket size')
]

choices = ['High-Value', 'At-Risk']

final['segment'] = np.select(conditions, choices, default='Neither')
```

```python
final['segment'].value_counts(normalize=True)*100
```

```python
segment_summary = final.groupby('segment').agg(
    customer_count=('user_id', 'count'),
    avg_basket_size=('avg_basket_size', 'median'),
    avg_order_cadence=('avg_days_bw_orders', 'median')
).reset_index()

segment_summary
```

# Inference:

At-Risk customers buy just as much per order as High-Value customers in terms of basket size — the only difference is they come back half as often. This is a frequency problem, not a basket size problem.

```python
# orders per month (30 days) --approx.
segment_summary['order_frequency'] = 30 / segment_summary['avg_order_cadence']
```

```python
segment_summary
```

```python
segment_summary['revenue_proxy'] = segment_summary['avg_basket_size'] * segment_summary['order_frequency']
segment_summary
```

# Revenue Proxy

```python
float(segment_summary[segment_summary['segment']=='High-Value']['revenue_proxy'].loc[1] / segment_summary[segment_summary['segment']=='At-Risk']['revenue_proxy'].loc[0] )
```

High-Value customers generate 2.7x the basket-frequency revenue proxy of At-Risk customers.

```python
reordered_data = final[final['segment']=='At-Risk'].merge(orders,on='user_id',how ='left')[['user_id','order_number','order_id']].merge(orders_prior,on='order_id',how='left')[['user_id','order_number','order_id','product_id','reordered']].merge(products,on='product_id',how='left').merge(departments,on='department_id',how='left')
```

```python
reordered_data.head()
```

```python
order1_depts = reordered_data[reordered_data['order_number'] == 1][['user_id', 'department']].drop_duplicates()

order3_depts = reordered_data[reordered_data['order_number'] == 3][['user_id', 'department']].drop_duplicates()

#We are comparing order 1 and order 3 here, because up until that point, all users on initially have been retained upto 3rd order. There is only drop off after the 3rd order
```

```python
order1_depts.head()
```

```python
order1_depts['user_id'].nunique() #116662
order3_depts['user_id'].nunique() #116662
```

```python
# Users who bought from each dept in order 1
order1_counts = order1_depts.groupby('department')['user_id'].nunique().reset_index(name='users_per_department_order1')
order1_counts
```

```python
both = order1_depts.merge(order3_depts, on=['user_id', 'department'], how='inner')
both
```

```python
both_counts = both.groupby('department')['user_id'].nunique().reset_index(name='users_reordered')
both_counts
```

```python
# Reorder rate on a department level
dept_reorder_rate = (order1_counts
    .merge(both_counts, on='department', how='left')
)
dept_reorder_rate['reorder_rate'] = dept_reorder_rate['users_reordered'] / dept_reorder_rate['users_per_department_order1']

# Top 5 leaking departments
dept_reorder_rate.sort_values('reorder_rate').head(5)
```

```python
# Similarly to compare High Value customers
reordered_data_hv = final[final['segment']=='High-Value'].merge(orders,on='user_id',how='left')[['user_id','order_number','order_id']].merge(orders_prior,on='order_id',how='left')[['user_id','order_number','order_id','product_id','reordered']].merge(products,on='product_id',how='left').merge(departments,on='department_id',how='left')

order1_depts_hv = reordered_data_hv[reordered_data_hv['order_number'] == 1][['user_id', 'department']].drop_duplicates()
order3_depts_hv = reordered_data_hv[reordered_data_hv['order_number'] == 3][['user_id', 'department']].drop_duplicates()

order1_counts_hv = order1_depts_hv.groupby('department')['user_id'].nunique().reset_index(name='users_order1')
both_hv = order1_depts_hv.merge(order3_depts_hv, on=['user_id', 'department'], how='inner')
both_counts_hv = both_hv.groupby('department')['user_id'].nunique().reset_index(name='users_reordered')

dept_reorder_rate_hv = order1_counts_hv.merge(both_counts_hv, on='department', how='left')
dept_reorder_rate_hv['reorder_rate_hv'] = dept_reorder_rate_hv['users_reordered'] / dept_reorder_rate_hv['users_order1']
```

```python
dept_reorder_rate_hv
```

```python
comparison = dept_reorder_rate.merge(
    dept_reorder_rate_hv[['department', 'reorder_rate_hv']],
    on='department',
    how='left'
)
comparison['gap'] = comparison['reorder_rate_hv'] - comparison['reorder_rate']
comparison.sort_values('gap', ascending=False).head(5)
```

# Inference:
----------------------------------------------------------------------------
- Personal care (13k users, 28.6%) — toothpaste, shampoo etc. are consumables people should be rebuying regularly. Low reorder rate here is a genuine signal that Instacart is losing that repeat purchase to somewhere else (Amazon, pharmacy). This is worth a targeted nudge
- Alcohol has the biggest gap (0.15) and a meaningful sample size (4,181 users)
----------------------------------------------------------------------------

- **Personal care** — lowest absolute reorder rate among At-Risk (28.6%)
- **Alcohol — largest** -  gap between At-Risk and High-Value (15 percentage points)
